# 내부 페이지 크롤링

에 뜨는 제품 이름, 가격, 제품타입, url 크롤링 후 csv 파일로 저장

```python
import asyncio
import pandas as pd
import csv
import re
from urllib.parse import urlparse, parse_qs, urlencode, urlunparse
from playwright.async_api import async_playwright

BASE = "https://www.oliveyoung.co.kr"
# 첫 페이지만 필요하므로 NUM_PAGES를 1로 설정하거나 반복문을 제거합니다.
SEARCH_URL = """https://www.oliveyoung.co.kr/store/search/getSearchMain.do?startCount=0&sort=RANK%2FDESC&goods_sort=WEIGHT%2FDESC%2CRANK%2FDESC&collection=ALL&realQuery=%EB%9D%BC%EC%9A%B4%EB%93%9C%EB%9E%A9%ED%8F%AC%EB%A7%A8&reQuery=&viewtype=image&category=&catename=LCTG_ID&catedepth=1&rt=&setMinPrice=&setMaxPrice=&listnum=24&tmp_requery=&tmp_requery2=&categoryDepthValue=1&cateId=10000010001&cateId2=&BenefitAll_CHECK=&query=%EB%9D%BC%EC%9A%B4%EB%93%9C%EB%9E%A9%ED%8F%AC%EB%A7%A8&selectCateNm=%EC%8A%A4%ED%82%A8%EC%BC%80%EC%96%B4+%EC%B9%B4%ED%85%8C%EA%B3%A0%EB%A6%AC%EC%97%90&firstTotalCount=16&typeChk=thum&branChk=&brandTop=&attrChk=&attrTop=&onlyOneBrand=&quickYn=N&cateChk=opened&benefitChk=&attrCheck0=&attrCheck1=&attrCheck2=&attrCheck3=&attrCheck4=&brandChkList=&benefitChkList=&_displayImgUploadUrl=https%3A%2F%2Fimage.oliveyoung.co.kr%2Fuploads%2Fimages%2Fdisplay%2F&recobellMbrNo=null&recobellCuid=8b47cf9f-efd1-48e4-8f83-10ee8a07945b&t_page=%EB%9D%BC%EC%9A%B4%EB%93%9C%EB%9E%A9%ED%8F%AC%EB%A7%A8%EA%B2%80%EC%83%89&t_click=%EC%83%81%ED%92%88%EB%B6%84%EB%A5%98%ED%95%84%ED%84%B0&t_search_name=&sale_below_price=&sale_over_price=&reChk=""".strip() # 브랜드에 맞게 뒷부분 바꿔야함

TARGET_BRAND = "라운드랩포맨"
UA = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"

def clean_product_name(name: str) -> str:
    """상품명에서 [단독기획], [1+1] 등 대괄호 및 내부 내용 삭제"""
    if not name: return ""
    # [ ] 로 감싸진 모든 텍스트를 제거하고 앞뒤 공백 정리
    cleaned = re.sub(r'\[.*?\]', '', name).strip()
    return cleaned

def classify_product_type(name: str) -> str:
    """수정된 분류 규칙: 에센스, 세럼, 앰플, 크림, 스킨, 토너, 미스트, 오일, 로션"""
    if not name: return "기타"
    n = name.lower()

    # 순서대로 매칭 (우선순위 고려 가능)
    rules = [
        ("에센스", [r"에센스", r"essence"]),
        ("세럼", [r"세럼", r"serum"]),
        ("앰플", [r"앰플", r"ampoule"]),
        ("크림", [r"크림", r"cream", r"밤\b"]),
        ("스킨", [r"스킨"]),
        ("토너", [r"토너", r"toner"]),
        ("미스트", [r"미스트", r"mist"]),
        ("오일", [r"오일", r"oil"]),
        ("로션", [r"로션", r"lotion", r"에멀전"]),
    ]

    for label, pats in rules:
        for pat in pats:
            if re.search(pat, n):
                return label
    return "기타"

async def collect_products_first_page(page):
    """검색 결과 첫 페이지만 수집"""
    await page.goto(SEARCH_URL, wait_until="domcontentloaded")
    await page.wait_for_timeout(2000)
    # 스크롤을 끝까지 내려야 가격 정보 등이 로드되는 경우가 있음
    await page.mouse.wheel(0, 3000)
    await page.wait_for_timeout(1000)

    # 상품 카드 선택자
    cards = page.locator("li:has(div.prd_info)")
    cnt = await cards.count()

    results = []
    for i in range(cnt):
        card = cards.nth(i)

        # 1. 브랜드 추출
        brand = await card.locator(".tx_brand").first.inner_text() if await card.locator(".tx_brand").count() > 0 else ""
        brand = brand.strip()

        # 타겟 브랜드가 아니면 패스
        if (TARGET_BRAND not in brand) and ('라운드랩' not in brand):
            continue

        # 2. 상품명 추출 및 정제
        raw_name = await card.locator(".tx_name").first.inner_text() if await card.locator(".tx_name").count() > 0 else ""
        name = clean_product_name(raw_name)

        # 3. 가격 추출 (현재 판매가 기준)
        price = "0"
        price_loc = card.locator(".tx_cur > span.tx_num").first
        if await price_loc.count() > 0:
            price = await price_loc.inner_text()
            price = price.replace(",", "").strip() # 숫자만 남김

        # 4. 링크 추출
        link = ""
        a_tag = card.locator("a[href*='goodsNo=']").first
        if await a_tag.count() > 0:
            href = await a_tag.get_attribute("href")
            link = href if href.startswith("http") else BASE + href

        # 5. 분류 적용
        ptype = classify_product_type(name)

        results.append({
            "brand": brand,
            "product_name": name,
            "product_type": ptype,
            "price": price,
            "product_link": link
        })

    return results

async def main():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context(user_agent=UA)
        page = await context.new_page()

        print(f"🔍 {TARGET_BRAND} 검색 첫 페이지 수집 시작...")
        total = await collect_products_first_page(page)

        await browser.close()

    # CSV 저장
    out = "/content/roundlab_product_list_fixed.csv"
    df = pd.DataFrame(total)
    df.to_csv(out, index=False, encoding="utf-8-sig")

    print(f"✅ 수집 완료: {out}")
    print(f"📊 총 수집 상품 수: {len(total)}개")
    print(df.head())

await main()
```

```python
import asyncio
import pandas as pd
from datetime import datetime
from playwright.async_api import async_playwright

async def extract_10_unique_date_reviews_simple():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
            viewport={'width': 1280, 'height': 1000}
        )
        page = await context.new_page()

        target_url = "https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000163494&dispCatNo=1000001000100160001,1000001000700070013&trackingCd=Result_1"
        start_date_limit = datetime(2025, 8, 1)
        results = []
        seen_dates = set()
        MAX_DAYS = 10

        try:
            await page.goto(target_url, wait_until="domcontentloaded")

            # 리뷰 탭 클릭
            await page.get_by_text("리뷰", exact=False).first.click()
            await page.wait_for_timeout(3000)

            # 최신순 정렬 클릭
            try:
                await page.locator("button.pc-sort-button").filter(has_text="최신순").click()
                await page.wait_for_timeout(3000)
            except: pass

            print(f"🔍 서로 다른 날짜 {MAX_DAYS}개를 찾을 때까지 탐색을 시작합니다...")

            retry_count = 0
            while len(results) < MAX_DAYS and retry_count < 20:
                review_items = page.locator("div.review-item, [class*='review-item']")
                item_count = await review_items.count()

                for i in range(item_count):
                    if len(results) >= MAX_DAYS: break

                    item = review_items.nth(i)
                    date_el = item.locator(".date, [class*='date']").first

                    if await date_el.count() > 0:
                        date_text = (await date_el.inner_text()).strip()

                        if date_text and date_text not in seen_dates:
                            try:
                                current_item_date = datetime.strptime(date_text, "%Y.%m.%d")
                                if current_item_date < start_date_limit: continue

                                # 1. 별점 추출
                                star_count = await item.evaluate("""(el) => {
                                    const stars = el.querySelectorAll('oy-review-star-icon');
                                    let c = 0;
                                    stars.forEach(s => {
                                        if (s.shadowRoot?.querySelector('path')?.getAttribute('fill') && s.shadowRoot.querySelector('path').getAttribute('fill') !== 'none') c++;
                                    });
                                    return c > 0 ? c + "점" : "";
                                }""")

                                # 🌟 [추가] 2. 스킨타입 추출 (지성, 민감성 등)
                                skin_elements = item.locator("dl.user_info_list span")
                                skin_count = await skin_elements.count()
                                skins = []
                                for j in range(skin_count):
                                    txt = (await skin_elements.nth(j).inner_text()).strip()
                                    if txt: skins.append(txt)
                                skin_info = ", ".join(skins) if skins else ""

                                # 3. 태그 수집
                                tags = []
                                tag_elements = item.locator("div.tag, .tag")
                                for j in range(await tag_elements.count()):
                                    tags.append(await tag_elements.nth(j).inner_text())

                                # 4. 리뷰내용 추출
                                content_el = item.locator("oy-review-review-content p, .content p").first
                                content = (await content_el.inner_text()).strip() if await content_el.count() > 0 else ""

                                seen_dates.add(date_text)
                                results.append({
                                    "날짜": date_text,
                                    "별점": star_count,
                                    "스킨타입": skin_info, # 결과에 추가됨
                                    "태그": ", ".join(tags) if tags else "",
                                    "리뷰내용": content
                                })
                                print(f"  ✅ [{len(results)}/10] 추가됨: {date_text} | {skin_info}")
                            except: continue

                if len(results) >= MAX_DAYS: break

                # 다음 페이지 시도
                await page.mouse.wheel(0, 2000)
                await page.wait_for_timeout(2000)

                try:
                    current_active_page_text = await page.locator(".pageing strong").inner_text()
                    next_page_num = int(current_active_page_text) + 1
                    await page.locator(f".pageing a:has-text('{next_page_num}')").first.click()
                    await page.wait_for_timeout(3000)
                except:
                    retry_count += 1
                    print("🔄 다음 페이지로 넘어가 데이터를 더 찾습니다...")

            if results:
                df = pd.DataFrame(results)
                df.to_csv("anua_10_days_reviews.csv", index=False, encoding="utf-8-sig")
                print(f"\n✨ 완료: {len(results)}개의 데이터를 저장했습니다.")
                print(df[['날짜', '스킨타입', '태그']].head())

        except Exception as e:
            print(f"⚠️ 에러: {e}")
        finally:
            await browser.close()

await extract_10_unique_date_reviews_simple()
```

### 전체 리뷰 추출
url 5개에 대해 리뷰 5개씩만 추출 -> csv 파일 만들기까지


이 코드를 수정해서 전체 제품에 대한 전체 리뷰 추출하시면 될겁니다!


```python
import asyncio
import pandas as pd
from datetime import datetime
from playwright.async_api import async_playwright
from tqdm.asyncio import tqdm

async def extract_roundlab_reviews_final():
    # 0. 설정 및 파일 로드
    try:
        product_df = pd.read_csv("/content/roundlab_product_list_fixed.csv")
        target_products = product_df
    except Exception as e:
        print(f"❌ 파일 읽기 실패: {e}")
        return

    # 설정값
    START_DATE = datetime(2025, 8, 1)
    END_DATE = datetime(2026, 1, 25)
    BIG_NUM = 1000000
    REVIEWS_PER_PRODUCT = BIG_NUM
    MAX_SCROLLS = 200  # 최대 스크롤 횟수 (무한루프 방지)

    final_rows = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
            viewport={'width': 1280, 'height': 1000}
        )

        pbar = tqdm(total=len(target_products), desc="🚀 제품별 수집 진행 중")

        for _, row in target_products.iterrows():
            url = row['product_link']
            base_info = {
                "product_name": row['product_name'],
                "product_type": row['product_type'],
                "price": row['price']
            }

            page = await context.new_page()
            current_product_count = 0
            collected_review_ids = set()  # 중복 방지용

            try:
                # 1. 페이지 이동
                await page.goto(url, wait_until="networkidle", timeout=60000)

                # 2. 리뷰 탭 클릭
                review_tab = page.get_by_text("리뷰", exact=False).first
                await review_tab.scroll_into_view_if_needed()
                await review_tab.click()
                await page.wait_for_selector(".review-item, [class*='review-item']", timeout=30000)

                # 3. 최신순 정렬 클릭
                sort_btn = page.locator("button.pc-sort-button").filter(has_text="최신순")
                if await sort_btn.count() > 0:
                    await sort_btn.click()
                    await page.wait_for_timeout(3000)

                # 4. 무한 스크롤로 리뷰 수집
                no_new_reviews_count = 0  # 새 리뷰가 없는 횟수
                stop_product = False

                for scroll_attempt in range(MAX_SCROLLS):
                    if current_product_count >= REVIEWS_PER_PRODUCT or stop_product:
                        break

                    # 현재 페이지의 모든 리뷰 아이템 가져오기
                    items = page.locator(".review-item, [class*='review-item']")
                    count = await items.count()

                    previous_count = current_product_count

                    for i in range(count):
                        if current_product_count >= REVIEWS_PER_PRODUCT:
                            break

                        item = items.nth(i)

                        # 중복 체크용 고유 ID 생성 (날짜 + 내용 일부)
                        try:
                            date_el = item.locator(".date, [class*='date']").first
                            if await date_el.count() == 0:
                                continue

                            date_text = (await date_el.inner_text()).strip()

                            # 리뷰 내용 미리 가져오기 (중복 체크용)
                            content_el = item.locator("oy-review-review-content p, .content p").first
                            content = (await content_el.inner_text()).strip() if await content_el.count() > 0 else "내용없음"

                            # 중복 체크
                            review_id = f"{date_text}_{content[:50]}"
                            if review_id in collected_review_ids:
                                continue

                            try:
                                curr_date = datetime.strptime(date_text, "%Y.%m.%d")

                                # 시작일 이전이면 수집 중단
                                if curr_date < START_DATE:
                                    print(f"   ⏱️ {date_text} 발견: 시작일 이전이므로 수집 중단.")
                                    stop_product = True
                                    break

                                # 날짜 범위 내의 리뷰만 수집
                                if START_DATE <= curr_date <= END_DATE:
                                    # 별점 추출
                                    star_count = await item.evaluate("""(el) => {
                                        const stars = el.querySelectorAll('oy-review-star-icon');
                                        let c = 0;
                                        stars.forEach(s => {
                                            const path = s.shadowRoot?.querySelector('path');
                                            if (path && path.getAttribute('fill') && path.getAttribute('fill') !== 'none' && path.getAttribute('fill') !== '#E5E5E5') c++;
                                        });
                                        return c + "점";
                                    }""")

                                    # 스킨타입 추출
                                    skin_info = "정보없음"
                                    try:
                                        skin_el = item.locator("dl.user_info_list span").first
                                        if await skin_el.count() > 0:
                                            skin_info = (await skin_el.inner_text()).strip()
                                    except:
                                        pass

                                    # 태그 수집
                                    tags = []
                                    tag_els = item.locator(".tag")
                                    tag_count = await tag_els.count()
                                    for j in range(tag_count):
                                        tags.append(await tag_els.nth(j).inner_text())
                                    tag_info = ", ".join(tags) if tags else "정보없음"

                                    # 데이터 합치기
                                    new_row = base_info.copy()
                                    new_row.update({
                                        "날짜": date_text,
                                        "별점": star_count,
                                        "스킨타입": skin_info,
                                        "태그": tag_info,
                                        "리뷰내용": content
                                    })
                                    final_rows.append(new_row)
                                    collected_review_ids.add(review_id)
                                    current_product_count += 1
                                    print(f"   ✅ {current_product_count}번째 리뷰 수집 완료 ({date_text})")

                            except ValueError:
                                continue

                        except Exception as e:
                            continue

                    if stop_product:
                        break

                    # 새 리뷰가 수집되지 않았는지 확인
                    if current_product_count == previous_count:
                        no_new_reviews_count += 1
                        if no_new_reviews_count >= 3:  # 3번 연속 새 리뷰 없으면 종료
                            print(f"   ⚠️ 더 이상 새로운 리뷰가 없습니다. (스크롤 {scroll_attempt + 1}회)")
                            break
                    else:
                        no_new_reviews_count = 0

                    # 스크롤 다운 (리뷰 영역 끝으로)
                    await page.evaluate("""() => {
                        const reviewSection = document.querySelector('.review-item, [class*="review-item"]')?.parentElement;
                        if (reviewSection) {
                            reviewSection.scrollIntoView({behavior: 'smooth', block: 'end'});
                        }
                        window.scrollBy(0, 800);
                    }""")

                    # 로딩 대기
                    await page.wait_for_timeout(2000)

                print(f"   📊 {base_info['product_name']}: 총 {current_product_count}개 리뷰 수집 완료")

            except Exception as e:
                print(f"⚠️ {base_info['product_name']} 진행 중 에러: {e}")
            finally:
                await page.close()
                pbar.update(1)

        await browser.close()
        pbar.close()

        # 5. 결과 저장
        if final_rows:
            result_df = pd.DataFrame(final_rows)
            result_df = result_df.drop_duplicates(subset=['product_name', '날짜', '리뷰내용'])
            result_df.to_csv("roundlab_comprehensive_reviews.csv", index=False, encoding="utf-8-sig")
            print(f"\n✨ 완료! 총 {len(result_df)}개의 리뷰를 저장했습니다.")
        else:
            print("❌ 수집된 데이터가 없습니다. 날짜 범위를 확인하세요.")

# 실행
await extract_roundlab_reviews_final()
```
